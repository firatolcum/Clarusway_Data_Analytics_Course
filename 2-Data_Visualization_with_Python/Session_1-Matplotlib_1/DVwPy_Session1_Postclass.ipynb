```python
divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"]
division_average_marks = [70, 82, 73, 65, 68]
boys_average_marks = [68, 67, 77, 61, 70]

index= np.arange(5)
width = 0.3

plt.bar(index, division_average_marks, width= width, color="green", label="Division Marks")
plt.bar(index+width, boys_average_marks, width=width, color="blue", label="Boys Marks")

plt.title("Horizontally Stacked Bar Graphs")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.xticks(index+ width/2, divisions)

plt.legend(loc="best")
plt.show()
```

```python
# FUNCTION METHOD
x = np.linspace(0, 10, 100)

plt.figure()

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.xlabel("x value")
plt.ylabel("sin value")
plt.title("Sinx Table")

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.xlabel("x value")
plt.ylabel("cos value")
plt.title("Cosx Table")

plt.subplot(2, 2, 3)
plt.plot(x, np.sin(x) / np.cos(x))
plt.xlabel("x value")
plt.ylabel("tan value")
plt.title("Tanx Table")

plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x) / np.sin(x))
plt.xlabel("x value")
plt.ylabel("cot value")
plt.title("Cotx Table")

plt.suptitle("Trigonometric Table")
plt.tight_layout()
plt.show()
```

```python
# OOP METHOD
fig, ax = plt.subplots(nrows=2, ncols=2)

x = [0, 30, 60, 90, 120, 150, 180]

ax[0,0].plot(x, np.sin(x))
ax[0,0].set_xlabel("x value")
ax[0,0].set_ylabel("Sin value")
ax[0,0].set_title("Sinx Table")

ax[0,1].plot(x, np.cos(x))
ax[0,1].set_xlabel("x value")
ax[0,1].set_ylabel("Cos value")
ax[0,1].set_title("Cosx Table")


ax[1,0].plot(x, np.tan(x))
ax[1,0].set_xlabel("x value")
ax[1,0].set_ylabel("Tan value")
ax[1,0].set_title("Tanx Table")

ax[1,1].plot(x, 1 / np.tan(x))
ax[1,1].set_xlabel("x value")
ax[1,1].set_ylabel("Tan value")
ax[1,1].set_title("Tanx Table")


plt.suptitle("Trigonometric Table")
plt.tight_layout()
plt.show()
```
