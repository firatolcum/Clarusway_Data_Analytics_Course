###  Lineplot

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
plt.figure(figsize=(20, 10), dpi=300)
sns.lineplot(x="year", y="passengers", data=flights);
```

```python
plt.figure(figsize=(20, 10), dpi=200)
sns.lineplot(x="month", y="passengers", data=flights);
```

```python
flights.groupby(["year", "month"]).sum()
```

```python
f_sum = flights.groupby(["year", "month"]).sum().reset_index()
f_sum.head()
```

```python
plt.figure(figsize=(20, 6))
sns.lineplot(x=f_sum.index, y=f_sum.passengers)
```

```python
plt.figure(figsize=(20, 6))
sns.lineplot(x=f_sum.year, y=f_sum.passengers, color="red")
```

```python
plt.figure(figsize=(20, 6))
sns.lineplot(x=f_sum.passengers, y=f_sum.year, color="red")
```

```python
flights.head()
```

```python
new_df = flights.pivot(index="year", columns="month", values="passengers")
new_df.head()
```

```python
plt.figure(figsize=(20, 10), dpi=300)
sns.lineplot(data=new_df);
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="year", y="passengers", hue="month", data=flights, palette="viridis")
```
