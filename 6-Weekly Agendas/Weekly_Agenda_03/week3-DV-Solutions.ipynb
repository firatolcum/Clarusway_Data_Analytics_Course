```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

**Use below dataframe for plots**

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

# 5. Unstuck the bars, annotate the hight of the bars on top of them

```python
labels=["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]

g = df.plot.bar(figsize=(9,6), rot=0)
g.set_xticklabels(labels)
for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.02, p.get_height()+0.5))
```

# 7. Plot the avg tip by size  (lineplot)

```python
tips.groupby('size').tip.mean().plot()
```

```python
sns.lineplot(data=tips, x="size", y="tip", ci=None)
```

```python
df1= tips.groupby('size')['tip'].mean()
df1
```

```python
fig, ax = plt.subplots()
ax.plot(df1.index, df1.values)
ax.set_xlabel("size");
```

# 9. Make a scatter plot between tip and total_bill

```python
tips.plot.scatter("tip", "total_bill")
```

```python
plt.scatter(tips["tip"], tips["total_bill"], linewidth = 0.001)
plt.xlabel("tip")
plt.ylabel("total_bill")
plt.show()
```

# 10. Set an additional dimension using size column

```python
tips.plot.scatter("tip", "total_bill", c="size", cmap="coolwarm")
```

```python
tips.plot.scatter(x = "tip", y = "total_bill", c = tips["size"], cmap = 'coolwarm')
```

```python
plt.scatter(tips["tip"], tips["total_bill"], linewidth = 0.001, c = tips["size"], cmap='coolwarm')
plt.colorbar()
plt.xlabel("tip")
plt.ylabel("total_bill")
plt.show()
```

```python
sns.scatterplot(x='tip', y='total_bill', hue='size', data= tips, palette='coolwarm',legend="full")
```

```python
# ncelikle scatter izdirme amacmz; iki deiken arasndaki ilikiyi ve dalm gzlemleyebilmektir.

# Seaborn ile Heatmap izdirirken cbar=False diyerek sadaki size skalasn kaldrabiliyoruz. 

# Ama seaborn ile scatterplot izdirirken cbar diye bir paremetre kullanamyoruz.

# Bunun yerine legend paremetresi ile oynayarak sa alt tarafta grdnz ekilde size stununu boyutsal olarak renk renk gsterebiliyor.
```
