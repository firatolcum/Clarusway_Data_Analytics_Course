### Seaborn Violinplot ###

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
cars = sns.load_dataset("mpg").dropna()
cars.head()
```

```python
cars.shape
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
cars.head()
```

```python
cars.cylinders.value_counts()
```

```python
sns.violinplot(cars.displacement);
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement);
```

```python
sns.violinplot(x="cylinders", y="displacement", data=cars);
```

```python
sns.violinplot(x="cylinders", y="displacement", data=cars, hue="origin");
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
              data=cars[cars.origin.isin(["japan", "europe"])])
plt.legend(loc=4)
plt.show()
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
              data=cars[cars.origin.isin(["japan", "europe"])],
              split=True)

plt.legend(loc=4)
plt.show()
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
              data=cars[cars.origin.isin(["japan", "europe"])],
              split=True,
              inner="quartiles")

plt.legend(loc=4)
plt.show()
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
              data=cars[cars.origin.isin(["japan", "europe"])],
              split=True,
              scale="count")

plt.legend(loc=4)
plt.show()
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
              data=cars[cars.origin.isin(["japan", "europe"])],
              split=True,
              scale="count",
              scale_hue=False)

plt.legend(loc=4)
plt.show()
```

```python
cars[cars.origin.isin(["japan", "europe"])].groupby("cylinders").origin.value_counts()
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement);
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              order=["japan", "europe", "usa"]);
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              linewidth=3);
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              bw=0.2);
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              bw=1);
```
