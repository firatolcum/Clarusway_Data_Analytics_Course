This E-commerce dataset is about the purchases of the customers for 1 year in UK. It demonstrates the customer's online buying habits.

**Attribute Information about "sales_082020" dataset:**

**``buyer_id:``** The distinguishing sequence of numbers or numbers and letters assigned by the department to a cardholder and appearing on that cardholder’s buyer identification card.<br>
**``buyer_country:``**  The country in which the Buyer’s Facilities are located.<br>
**``seller_id:``** A unique alphanumeric code assigned by FPX Operator to identify the Seller.<br>
**``seller_country:``** The country in which the Seller’s Facilities are located.<br>
**``product_code:``** A unique identifier, assigned to each finished/manufactured product which is ready, to be marketed or for sale.<br>
**``category:``** The particular group of related products.<br>
**``brand:``** A type of product manufactured by a particular company under a particular name.<br>
**``currency:``** The money in circulation in the given country.<br>
**``price:``** The price of each product which were sold at a given date.<br>
**``date:``** Invoice Date, the day when each transaction was generated at a given date.<br>


**Attribute Information about "currency_rates_082020_EUR" dataset:**

**``currency:``** The money in circulation in the given country.<br>
**``date:``** The currency exchange date, the day when each exchange transaction was generated.<br>
**``rate:``** The currency exchange rate when each transaction was generated at a given date.<br>

### <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">Task 07: Which brand had the largest absolute € difference in average transaction value between domestic and international?</p>

<a id="2"></a>

style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

**Let's remember our DataFrame first.**

```python
df_merged.head(3)
```

**First we create a new feature named "Sale_Type" including the values of "Domestic Sale" and  "International Sale".**

```python
df_merged['Sale_Type'] = df_merged.apply(lambda x: "Domestic Sale" if x['buyer_country'] == x['seller_country'] else "International Sale", axis=1)
df_merged.sample(10)
```

**Let's check if there have been any "Null" value in the new feature.** 

```python
df_merged['Sale_Type'].value_counts(dropna=False)
```

**Let us calculate the average sales according to "brand" & " Sale_Type"**

```python
df_merged.groupby(['brand', 'Sale_Type'])["total_Euro"].mean()
```

**Let's create a new feature named "Brand_Average_Sale" demonstrating the average sales according to "brand" & " Sale_Type"**

```python
df_merged["Brand_Average_Sale"] = df_merged.groupby(['brand', 'Sale_Type'])["total_Euro"].transform("mean")
df_merged["Brand_Average_Sale"].fillna("Brand Not-Applicable", inplace=True)
```

### Generic Solution For Task 07

```python
sales_domestic = (
                    df_merged
                    .groupby(['brand', 'buyer_country', 'seller_country'])
                    .agg(avg_sales=('total_Euro', 'mean'))
                    .round(2)
                    .query('buyer_country == seller_country')
                    .reset_index(level='brand')
                    .reset_index(drop=True)
                    .groupby('brand')
                    .agg(avg_sales=('avg_sales', 'mean'))
                )
```

```python
sales_domestic
```

```python
sales_international = (
                        df_merged
                        .groupby(['brand', 'buyer_country', 'seller_country'])
                        .agg(avg_sales=('total_Euro', 'mean'))
                        .round(2)
                        .query('buyer_country != seller_country')
                        .reset_index(level='brand')
                        .reset_index(drop=True)
                        .groupby('brand')
                        .agg(avg_sales=('avg_sales', 'mean'))
                       )
```

```python
sales_international
```

```python
sales_dom_int = sales_international.merge(sales_domestic, how='inner',
                                          left_index=True, right_index=True,
                                          suffixes=('_international', '_domestic'))
```
