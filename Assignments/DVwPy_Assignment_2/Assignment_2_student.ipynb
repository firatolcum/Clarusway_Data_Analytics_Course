## 2: Plot 12 most dangerous cities

```python
kill.head(3)
```

```python
city_freq = kill.city.value_counts().head(12)
city_freq
```

```python
# with seaborn barplot
fig, ax =plt.subplots(figsize=(8,5))

ax = sns.barplot(x=city_freq.index, y=city_freq.values)

ax.set_title("Most dangerous cities", color="blue", size=16)
ax.tick_params(axis='x', labelrotation = 45)
# plt.xticks(rotation=90)  # easy way for rotation
```

```python
# with seaborn countplot
fig, ax =plt.subplots(figsize=(8,5))

sns.countplot(kill.city, order=kill.city.value_counts().head(12).index)

ax.set_title("Most dangerous states", color="blue", size=15)
ax.tick_params(axis="x",labelrotation=45)
# plt.xticks(rotation=90)  # easy way for rotation
```
